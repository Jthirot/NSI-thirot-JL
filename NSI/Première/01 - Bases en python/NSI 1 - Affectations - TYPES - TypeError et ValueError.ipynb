
<p style="padding:5px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:2em;padding:15px;">I. AFFECTATION</p>


<p style="padding:5px;background-color:#bb6666;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;"> A RETENIR</p>

Lorsque l'on écris en python :
```python
ma_variable = 1
```

ou bien avec une expression dans le membre de droite :
```python
ma_variable = 2*7
```

Si **ma_variable** n'existe pas déjà, on fait 3 choses :

1. Il évalue le membre de droite. (ici, il vaut trivialement 1 dans le 1er exemple, ou 14 dans le second)
2. dans un emplacement de la mémoire, il écrit la valeur. Cet emplacement est repréré par une adresse, par exemple, l'adresse 3005392 qui est un simple numéro qui situe un emplacmeent dans la mémoire.
+ il crée aussi une 
<span style="padding:1px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;"> variable</span> nommée **ma_variable** qui est un mot utilisable dans le code, qui réfère à l'emplacement mémoire.

Si au contraire, **ma_variable** existe déjà :

1. Il évalue le membre de droite.
2. dans l'emplacement de la mémoire correspondant à **ma_variable**, il écrit la valeur. 

Ce n'est pas python qui choisit l'emplacement mémoire, c'est le système d'exploitation qui gère cela. En fait python envoie une requête à l'OS pour stocker une valeur, l'OS choisit un emplacement vide, stocke la valeur et renvoie à l'interpréteur python l'adresse mémoire utilisée.
<br><br>

<div style="padding:5px;background-color:#bb6666;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;"> DEFINITION
    
une affectation est faite dans toutes les instructions du type var = expression
    
Cela consiste à :
    
    
+ Evaluer le membre de droite
+ stocker le résultat dans un emplacement mémoire
+ Cette valeur sera par la suite accessible en utilisant le nom de la variable
</div>

Pas abus de langage, on dit que la variable **prend la valeur** de l'expression.

par exemple :
```python
a = 1
```
on dit que **a** prend la valeur **1**.

````python
# une affectation :
ma_variable = 1

# afficher l'adresse mémoire :
print(id(ma_variable))
````

````python
# plusieurs affectation :
x = 1
y = 3
somme = x + y

# afficher les adresses mémoire :
print( "x est stocké à l'adresse :", id(x) )
print( "y est stocké à l'adresse :", id(y) )
print( "somme est stocké à l'adresse :", id(somme) )
````

Il existe différents <span style="padding:1px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;">types</span> de valeur, et on doit toujours veiller à ne pas mélanger des choux avec des carotes.

On sait facilement calculer 1 + 3 ou même 1 + 3.5 mais on ne sait que faire de 1 + "Albert"

Parce que 1 est un **nombre entier**, et "Albert" est une **chaine de caractère**.

Pour ce début d'année, il faut avoir à l'esprit qu'il existe 4 types de base :

<p style="padding:5px;background-color:#bb6666;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;"> A RETENIR</p>

|type|description|
|--|--|
|int|les nombres entiers|
|float|les nombres avec une virgule|
|str|les chaînes de caractères|

**int** est l'abréviation de **interger** qui ignifie entier en anglais

**float** est l'abbréviation de **floating point** qui se traduit en français par **nombre à virgule flottante**. Nous aurons l'occasion plus tard de reparler de ce nom étrange, pour le moment considérez qu'il s'agit de nombres décimaux.

**str** est l'abréviation de **string**, et désigne les chaine de charactères (character strings).

**bool** est l'abréviation de **boolean**, soit booléen en français. C'est un type moins intuitif mais très important en informatique. Une variable booléenne ne peut avoir que 2 valeurs : **True** ou **False** (Vrai ou Faux)

Voyons quelques exemples :


<p style="padding:5px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;padding:15px;">II.2 différence entre <b>int</b> et <b>float</b></p>

````python
un_entier = 1
print( "le type de la variable un_entier est :",type(un_entier))

mot = "Albert"
print( "le type de la variable mot est :",type(mot))

un_float = 1.0
print( "le type de la variable un_float est :",type(un_float))

bool = True
print( "le type de la variable bool est :",type(bool))
````

Vous vous posez peut-être la question de savoir si 1 et 1.0 sont identiques ? 

En vérité ils ne le sont pas, puisqu'ils n'ont pas le même type. Cela veux dire qu'il ne sont pas représenté de la même façon écriture binaire. Néanmoins, ils ont bien la même valeur.

Illustrons cela en utilisant le test d'égalité **==** et le test d'identité **is**.

+ Le premier, a == b est True si a et b on la même valeur.
+ Le second, a is b est True si a et b sont identiques.

````python
a = 1
b = 1.0
print( "a est égal à b ? -> ",a == b)

print( "a est identique à b ? -> ",a is b)
````


<p style="padding:5px;background-color:#bb6666;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;"> A RETENIR</p>

<p style="padding:15px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;padding:15px;">II.3 les conversions de types</p>
Il est possible, et souvent on a besoin de la faire, de modifier le type d'une valeur.

Par exemple :

````python
# on défini une variable
a = 1
print( "a vaut :", a, "le type de a est : " , type(a) )

# on converti a en float :
a = float(a)
print( "a vaut :", a, "le type de a est : " , type(a) )
````

````python
# on défini une variable
a = 1.0
print( "a vaut :", a, "le type de a est : " , type(a) )

# on converti a en float :
a = int(a)
print( "a vaut :", a, "le type de a est : " , type(a) )
````

````python
# Mais ATTENTION :
a = 1.73
print( "a vaut :", a, "le type de a est : " , type(a) )

# on converti a en float :
a = int(a)
print( "a vaut :", a, "le type de a est : " , type(a) )
````

Dans ce dernier exemple, 1.73 a été converti en **int** mais les chiffres après la virgule sont omis.

De même, avec un nombre négatif :

````python
# Mais ATTENTION :
a = -1.73
print( "a vaut :", a, "le type de a est : " , type(a) )

# on converti a en float :
a = int(a)
print( "a vaut :", a, "le type de a est : " , type(a) )
````

Une telle conversion n'est pas toujours possible, il faut que cela ait du sens :

````python
# on défini une variable
a = "albert"
print( "a vaut :", a, "le type de a est : " , type(a) )

# on converti a en int :
a = int(a)
print( "a vaut :", a, "le type de a est : " , type(a) )
````

Voila un second type d'erreur que vous rencontrerez souvent : **ValueError**

Le message invalid literal for int() with base 10: 'albert' se traduit par : la valeur litérale pour la fonction int() n'est pas valide en base 10 : 'albert'

En clair : vous avez essayez de convertir en int une valeur qui ne peut pas être interprétée comme un entier en base 10.

 <span style="padding:1px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;">affectation</span> 
 
Une affectation est une instruction de type : 
```python
var = expression
```
L'expression à droite du **=** est d'abord évalué, puis le résultat est écrit en mémoire, et le nom de la variable permet de réutiliser cette valeur.



 <span style="padding:1px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;">expression</span> 
 
Une expression n'est pas à proprement parler une instruction, c'est quelques chose que python peut évaluer (c'est a dire donner sa valeur) : 

Par exemple, dans :
```python
var = expression
```
expression peut être un nombre, un calcul, une chaine de caractères, un bouléen ou des opérations plus complexes que nous verrons plus tard.

Notez qu'on parle d'expression booléenne par exemple : 
```python
a == 3
```
est une expression booléenne. et donc l'instruction :
```python
b = a == 3
```
Est l'affectation à la variable b du résultat de l'expression boolénne a == 3

 <span style="padding:1px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;">types de base</span> 

|type|description|
|--|--|
|int|les nombres entiers|
|float|les nombres avec une virgule|
|str|les chaînes de caractères|
|bool|les booléens|

On peut convertir des float en int, ou inversement. On peut également convertir des chaine **str** en **int** ou en **float** mais seulement si la chaine contient un nombre compréhensible.

Exemple : float(1.23)

 <span style="padding:1px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;">Messages d'erreur</span> 

|type|description|
|--|--|
|ValueError|on utilise une valeur non valide dans une expression. exemple : int("albert")|
|SyntaxError|l'instruction n'a pas de sens, par exemple il manque une ) ou toute autre syntaxe erronée|
|TypeError|on a fait une opération entre 2 types incompatibles, exemple : 1 + "albert"|
|NameError|on utilise, dans une expression, une variable qui n'existe pas (non définie)|
