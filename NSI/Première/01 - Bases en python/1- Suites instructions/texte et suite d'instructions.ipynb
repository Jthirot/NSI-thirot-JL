Cette erreur, lors de l'exécution, est fréquente. 
    
<b>Vous devez lire les messages d'erreurs et apprendre à les comprendre pour être en mesure de corriger vos codes.</b>

**Exemples :**
    
<span style="padding-left:10px;color:red;">SyntaxError : line 10 :  unexpectedEOF while parsing</span>

Vous avez oubié une parenthèse (ou crochet ou accolade) à la ligne qui précède la ligne 10. 
    
**parsing** : analyse. 
    
exemple :
    
    ligne 9 : print("Bonjour"
    -> L'erreur appraitra ligne 10
    
<span style="padding-left:10px;color:red;">SyntaxError : line 10 :  while scanning string litteral</span>

Vous avez oubié une guillement. Python signale qu'une chaine de caractères n'est pas terminée. par exemple :
    
    print("Bonjour)
    

    
<span style="padding-left:10px;color:red;">SyntaxError : line 10 :  invalid syntax</span>

Vous avez écrit une instruction que python ne sait pas interpréter. Il existe un très grand nombre de cas, il faut relire la ligne et trouver ce qui ne va pas.

exemple :
    
    print(Bonjour monsieur)
    
    Bonjour monsieur n'est ni une chaine de caratères (pas de guillemet) ni une variable (deux mots).
    
autre exemple :
    
    print(ma_var1 ma_var2)
    
    Sans doute ma_var1 et ma_var2 sont des variables duments définies, mais il faudrait une virgule pour que cette instruction soit valide.
    
<span style="padding:1px;padding-left:10px;padding-right:10px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1em;">NameError</span> 
    
Erreur également très fréquente, vous devez **absolument** réagir sans hésiter : une variable est utilisée mais n'a pas été définie.
    
exemple :
    
    
    
    print(a)
    
    NameError : ligne 10 : name 'a' is not defined
    
    signifie que vous demandez d'afficher la valeur de a, mais il n'y a pas de variable nommée a dans votre code.
