La fonction draw() est exécutée en boucle. Cela permet de raffraichir en permanence le dessin. 

Pour le moment nous allons simplement dessiner quelques objets dont nous aurons besoin. A savoir :

- la fonction **square(x,y,taille)** dessinne des carrés. x et y sont les coordonnées du coin suppérieur gauche du carré, et taille est la longeur des cotés.
- la fonction **rect(x,y,width,height)** dessinne des rectangle. x et y sont les coordonnées du coin suppérieur gauche du rectangle, width est la largeur et height est la longeur du rectangle.
- la fonction **circle(x,y,radius)** dessinne des cercles. x et y sont les coordonnées du centre du cercle, radius est le rayon.
- text(texte,x,y) affiche la chaine de caractère texte (str) aux coordonnées x,y. Par defaut, c'est le coin inférieur droit du texte qui est positionné en (x,y) mais il est possible de modifier ce comportement pour écrire centré (horizontalement ou verticalement), justifié à droite ou à gauche etc..

La façon dont les éléments sont dessinés peut être modifié, en précédent les appels par les appels de **textSize()**, **fill()** et/ou **stroke()**
- la fonction **textSize(color)** permet de fixer la taille du texte. color est une couleur, qui peut être le nom de la couleur (ex: "white"), ou un triplet de 3 nombres (ex : (200,120,255) ) ou une chaine hexadecimale (ex: "#ffaacc") voir même un entier, dans ce cas c'est un niveau de gris (ex: 200)
- la fonction **fill()** définie la couleur de remplissage utilisée, et peut être appelé avant chaque élément.
- la fonction **stroke()** définie la couleur des bordures.


Les coordonnées sont données en nombres de pixels. (x,y) représente un point du dessin. 

<div style="margin:10px;background:#bb6666;color:white">

    </li></ul>

Par exemple, dans un canvas de 200x200 pixels, le centre sera aux coordonnées (100,100)



````python
from p5 import *

def setup() :
    createCanvas(200, 200) # on cré une fenêtre de 100x100 pixels
    background("#99cccc")  # on choisi une couleur pour le fond

def draw() :
    # dessinons 2 carrés blanc de coté 20 à (10,10) et un bleu à (50,10) :
    fill("white")
    square(10,10,20) # dessine un carré blanc de coté 20 aux coordonnées (10,10)
    fill(0,200,0)
    square(50,10,40) #dessine un carré vert de coté 40 aux coordonnées (50,10)

    # dessinons un rectangle gris, de dimensions 20x30, à 10,40
    fill(100)
    rect(10,40,20,30)
    
    # nous voulons ajoter un cercle blanc au centre de ce rectangle.
    # le centre du rectangle est :
    # en abcisse  : 
    # le bord gauche est à 10, la longueur est 20 -> le centre est à 20
    # en ordonnées : 
    # le bord supérieur est à 40, la hauteur 30 -> le centre est à 55
    
    # dessinons un cercle de rayon 15 au centre du rectangle
    fill("white")
    circle(20,55,15)
    
    # ajoutons un texte :
    textSize(18)
    fill("black")
    text("un texte",10,100)
    
# lancement :
run()
````

````python
# setup et draw

def setup() :
    createCanvas(200, 200) # on cré une fenêtre de 100x100 pixels
    background("#99cccc")  # on choisi une couleur pour le fond

def draw() :
    # dessinons 2 carrés blanc de coté 20 à (10,10) et un bleu à (50,10) :
    fill("white")

    square(10,10,20) # dessine un carré blanc de coté 20 aux coordonnées (10,10)
    fill(0,200,0)
    square(50,10,40) #dessine un carré vert de coté 40 aux coordonnées (50,10)

    # dessinons un rectangle gris, de dimensions 20x30, à 10,40
    fill(100)
    rect(10,40,20,30)
    
    # nous voulons ajoter un cercle blanc au centre de ce rectangle.
    # le centre du rectangle est :
    # en abcisse  : 
    # le bord gauche est à 10, la longueur est 20 -> le centre est à 20
    # en ordonnées : 
    # le bord supérieur est à 40, la hauteur 30 -> le centre est à 55
    
    # dessinons un cercle de rayon 15 au centre du rectangle
    # NOUS AJOUTONS stroke POUR AVOIR UNE borDURE JAUNE
    fill("white")
    stroke("yellow")  # <- on défini la couleur de la bordure
    circle(20,55,15)
    stroke("black")  # <- on redéfini la couleur de la bordure en noir pour les éléments suivants
    
    # ajoutons un texte :
    textSize(18)
    fill("black")
    text("un texte",10,100)
    
# lancement :
run()
````

Vous remarquez que le texte est écris en gras... en fait stroke dessine la bordure noire aussi sur les éléments textes.
Pour éviter cela, on peut préférer utiliser la fonction **noStroke()** :

````python
from p5 import *

def setup() :
    createCanvas(200, 200) # on cré une fenêtre de 100x100 pixels
    background("#99cccc")  # on choisi une couleur pour le fond

def draw() :
    # dessinons 2 carrés blanc de coté 20 à (10,10) et un bleu à (50,10) :
    fill("white")

    square(10,10,20) # dessine un carré blanc de coté 20 aux coordonnées (10,10)
    fill(0,200,0)
    square(50,10,40) #dessine un carré vert de coté 40 aux coordonnées (50,10)

    # dessinons un rectangle gris, de dimensions 20x30, à 10,40
    fill(100)
    rect(10,40,20,30)
    
    # nous voulons ajoter un cercle blanc au centre de ce rectangle.
    # le centre du rectangle est :
    # en abcisse  : 
    # le bord gauche est à 10, la longueur est 20 -> le centre est à 20
    # en ordonnées : 
    # le bord supérieur est à 40, la hauteur 30 -> le centre est à 55
    
    # dessinons un cercle de rayon 15 au centre du rectangle
    # NOUS AJOUTONS stroke POUR AVOIR UNE borDURE JAUNE
    fill("white")
    stroke("yellow")  # <- on défini la couleur de la bordure
    circle(20,55,15)
    noStroke()  # <- on supprime la définition de  la couleur des bordures.
    
    # ajoutons un texte :
    textSize(18)
    fill("black")
    text("un texte",10,100)
    
# lancement :
run()
````

<p style="padding:5px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:2em;">5. Detecter un clic</b></span>

Jusqu'à présent, nos dessins sont statiques.

**Draw()** est appelée en boucle, mais elle fait à chaque fois le même dessin, donc on ne vois rien évoluer.

Nous allons ajouter une fonction clic dans notre code. Dans cette fonction nous utiliserons une autre **variable globale** : 

+ **mouseIsPressed**  vaut True quand un bouton de la souris est enfoncée et False sinon.

A chaque exécution de **draw()** la valeur de mouseIsPressed est évalué automatiquement. 

Pour le moment nous ne ferons qu'afficher dans la console un message si le bouton est enfoncé.

### Lancez le code ci-dessous, puis effectuez quelques clics en regardant les messages dans la console

Faites des clics aussi rapides que possible, et constatez que à chaque fois, la fonction **clic** a été exécutée plusieur fois...

````python
from p5 import *

def preload() :
    global FLECHE_JAUNE
    # on définie la variable fleche_jaune qui contiendra un objet image
    FLECHE_JAUNE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-34-01/flehce_jaune.jpg")
    
def setup() :
    createCanvas(400, 200) # on cré une fenêtre de 100x100 pixels
    background("#99cccc")  # on choisi une couleur pour le fond

def clic() :
    if mouseIsPressed :
        print("bouton de la souris enfonce")
        
def draw() :
    decal_x = 20
    decal_y = 50
    large = 40
    haut = 56
    for i in range(5) :
        # on dessine l'image (en utilisant la variable **globale** FLECHE_JAUNE)
        image(FLECHE_JAUNE,decal_x + i*large,decal_y)
    
    clic()
    
# lancement :
run(preload=preload)
````

<span style="padding:5px;background-color:#66bb66;border:1px solid black;border-radius:20px;font-weight:bold;color:white;font-size:1.2em;">Pour traiter les clics un par un</b></span>

Il n'est pas très satisfaisant qu'un seul clic génère plusieurs fois l'exécution de notre fonction **clic()**. Les appels successifs de draw() sont trop rapides pour nous !!!

Nous allons utiliser un **drapeau** : c'est une variable, généralement booléenne, qui va nous permettre de savoir si un clic a lieu.

+ Elle sera initialisée dans le programme principal.
+ Elle sera utilisée dans clic, et devra donc être globale.

### lancez le code, et cliquez dans l'image

vous pouvez faire des clics long, la fonction clic sera exécutée à chaque appel de **draw()** mais elle ne fera le travail qu'un fois à chaque clic.

````python
from p5 import *

def preload() :
    global FLECHE_JAUNE
    # on définie la variable fleche_jaune qui contiendra un objet image
    FLECHE_JAUNE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-34-01/flehce_jaune.jpg")
    

def setup() :
    createCanvas(400, 200) # on cré une fenêtre de 100x100 pixels
    background("#99cccc")  # on choisi une couleur pour le fond

def clic() :
    global CLIC_EN_COURS
    
    if mouseIsPressed :
        if not CLIC_EN_COURS :          # si un clic n'est pas deja en cours
            CLIC_EN_COURS = True        # alors on met clic_en_cours à True pour bloquer d'autre exécution
                                
            print("bouton de la souris enfonce")
    else :
        CLIC_EN_COURS = False           # le bouton est relaché, on remet clic_en_cours à False
        
def draw() :
    decal_x = 20
    decal_y = 50
    large = 40
    haut = 56
    for i in range(5) :
        # on dessine l'image (en utilisant la variable **globale** FLECHE_JAUNE)
        image(FLECHE_JAUNE,decal_x + i*large,decal_y)
        
    
    clic()
    
# lancement :
CLIC_EN_COURS = False
run(preload=preload)
````

### Le script ci-dessous va permettre d'afficher les coordonnées du clic


````python
from p5 import *

def preload() :
    global FLECHE_JAUNE
    # on définie la variable fleche_jaune qui contiendra un objet image
    FLECHE_JAUNE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-34-01/flehce_jaune.jpg")
    
def setup() :
    createCanvas(400, 200) # on cré une fenêtre de 100x100 pixels
    background("#99cccc")  # on choisi une couleur pour le fond

def clic() :
    global CLIC_EN_COURS
    global LARGE,HAUT,DECAL_X,DECAL_Y,NB_IMAGES
    
    if mouseIsPressed :
        if not CLIC_EN_COURS :          # si un clic n'est pas deja en cours
            CLIC_EN_COURS = True        # alors on met clic_en_cours à True pour bloquer d'autre exécution
                                
            print("clic détecté aux coordonnées :",mouseX,mouseY)
    else :
        CLIC_EN_COURS = False           # le bouton est relaché, on remet clic_en_cours à False
        
def draw() :

    for i in range(NB_IMAGES) :
        # on dessine l'image (en utilisant la variable **globale** FLECHE_JAUNE)
        image(FLECHE_JAUNE,DECAL_X + i*LARGE,DECAL_Y)
        
    
    clic()
    
####################################################
###                PROGRAMME PRINCIPAL          ####
####################################################

# largeur/hauteur des images :
LARGE = 40
HAUT = 56
NB_IMAGES = 7

# position des images 
DECAL_X = 50
DECAL_Y = 50

# lancement :
CLIC_EN_COURS = False
run(preload=preload)
````

````python
from module_part4 import *
from p5 import *

def preload() :
    global FLECHE_JAUNE,FLECHE_ROUGE
    # on définie la variable fleche_jaune qui contiendra un objet image
    FLECHE_JAUNE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-34-01/flehce_jaune.jpg")
    FLECHE_ROUGE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-19-31/flehce_rouge.jpg")

def setup() :
    createCanvas(300, 400) # on cré une fenêtre de 100x100 pixels
    background("#ivory")  # on choisi une couleur pour le fond

def image_cliquee(x,y) :
    """
    IN : x et y : coordonnées du clic
    OUT :
        - si on a cliqué dans la zone des images : la fonction renvoie le numéro de l'image cliquée
        - Sinon : elle renvoie -1
    """
    global LARGE,HAUT,DECAL_X,DECAL_Y,NB_IMAGES
    
    if DECAL_X <= x <= DECAL_X + NB_IMAGES * LARGE and DECAL_Y <= y <= DECAL_Y + HAUT :
        return int(x-DECAL_X)//LARGE
    return -1

def clic() :
    global CLIC_EN_COURS
    global JOUEUR,GRILLE
   
    if mouseIsPressed :
        if not CLIC_EN_COURS :          # si un clic n'est pas deja en cours
            CLIC_EN_COURS = True        # alors on met clic_en_cours à True pour bloquer d'autre exécution
            num = image_cliquee(mouseX,mouseY)
            if num >= 0 :
                GRILLE = joue_jeton(GRILLE,JOUEUR,num)
                JOUEUR = 3 - JOUEUR
    else :
        CLIC_EN_COURS = False           # le bouton est relaché, on remet clic_en_cours à False

def affiche_grille(x,y) :
    """
    Dession une grille constituée de 7 lignes, contenant chacune 7 carrés.
    La couleur des carrés est (0,0,200), leur taille est indiquée par la variable globale LARGE.
    Le premier carré est déssiné aux coordonnées x,y reçues en paramètres.
    """
    global LARGE,HAUT
    
    for ligne in range(6) :
        for colonne in range(7) :
            fill(0,0,200)
            square(x+colonne*LARGE, y+ligne*LARGE, LARGE)
            pion(x+colonne*LARGE , y+ligne*LARGE, ligne, colonne)
    
def pion(x,y,i,j) :
    global GRILLE,LARGE
    centre_x = x + LARGE//2
    centre_y = y + LARGE//2
    if GRILLE[i][j] == 0 : color="white"
    elif GRILLE[i][j] == 1 : color="red"
    else : color="yellow"
    
    fill(color)
    circle(centre_x,centre_y,LARGE * 3 //4,)
    
def draw() :
    global LARGE,DECAL_X,DECAL_Y,NB_IMAGES
    global FLECHE_JAUNE,FLECHE_ROUGE,JOUEUR
    
    if JOUEUR == 1 :
        fleche = FLECHE_ROUGE
    else :
        fleche = FLECHE_JAUNE
        
    for i in range(NB_IMAGES) :
        # on dessine l'image
        image(fleche, DECAL_X + i*LARGE,DECAL_Y)
        
    affiche_grille(DECAL_X,DECAL_Y+56)
    
    clic()
    
####################################################
###                PROGRAMME PRINCIPAL          ####
####################################################

# initialisation de la grille
GRILLE = creation_grille_vierge()

# largeur/hauteur des images :
LARGE = 40
HAUT = 56
NB_IMAGES = 7

# position des images 
DECAL_X = 10
DECAL_Y = 40

# lancement :
JOUEUR = 1
CLIC_EN_COURS = False

run(preload=preload)
````
