````python
##################################
#####   Imports de modules #######
##################################
from module_part4 import *
from p5 import *


####################################
#####   Fonctions et procédures ####
####################################

def preload() :
    global FLECHE_JAUNE,FLECHE_ROUGE
    # on définie la variable fleche_jaune qui contiendra un objet image
    FLECHE_JAUNE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-34-01/flehce_jaune.jpg")
    FLECHE_ROUGE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-19-31/flehce_rouge.jpg")

def setup() :
    createCanvas(300, 400) # on cré une fenêtre de 100x100 pixels
    background("#ivory")  # on choisi une couleur pour le fond

def image_cliquee(x,y) :
    """
    IN : x et y : coordonnées du clic
    OUT :
        - si on a cliqué dans la zone des images : la fonction renvoie le numéro de l'image cliquée
        - Sinon : elle renvoie -1
    """
    global LARGE,HAUT,DECAL_X,DECAL_Y,NB_IMAGES
    
    if DECAL_X <= x <= DECAL_X + NB_IMAGES * LARGE and DECAL_Y <= y <= DECAL_Y + HAUT :
        return int(x-DECAL_X)//LARGE
    return -1


def affiche_grille(x,y) :
    """
    Dession une grille constituée de 7 lignes, contenant chacune 7 carrés.
    La couleur des carrés est (0,0,200), leur taille est indiquée par la variable globale LARGE.
    Le premier carré est déssiné aux coordonnées x,y reçues en paramètres.
    """
    global LARGE,HAUT
    
    for ligne in range(6) :
        for colonne in range(7) :
            fill(0,0,200)
            square(x+colonne*LARGE, y+ligne*LARGE, LARGE)
            pion(x+colonne*LARGE , y+ligne*LARGE, ligne, colonne)
    
def pion(x,y,i,j) :
    global GRILLE,LARGE
    centre_x = x + LARGE//2
    centre_y = y + LARGE//2
    if GRILLE[i][j] == 0 : 
        fill("white")
        circle(centre_x,centre_y,LARGE * 3 //4,)
    elif GRILLE[i][j] == 1 : 
        fill("red")
        circle(centre_x,centre_y,LARGE * 3 //4,)
        stroke(200,0,0)
        circle(centre_x,centre_y,LARGE * 3 //5)
        circle(centre_x,centre_y,LARGE * 3//10)
        noStroke()
    else : 
        fill("yellow")
        circle(centre_x,centre_y,LARGE * 3 //4,)
        stroke(200,200,0)
        circle(centre_x,centre_y,LARGE * 3 //5)
        circle(centre_x,centre_y,LARGE * 3//10)
        noStroke()

def clic() :
    global CLIC_EN_COURS
    global JOUEUR,GRILLE
   
    if mouseIsPressed :
        if not CLIC_EN_COURS :          # si un clic n'est pas deja en cours
            CLIC_EN_COURS = True        # alors on met clic_en_cours à True pour bloquer d'autre exécution
            num = image_cliquee(mouseX,mouseY)
            if num >= 0 :
                GRILLE = joue_jeton(GRILLE,JOUEUR,num)
                JOUEUR = 3 - JOUEUR
    else :
        CLIC_EN_COURS = False           # le bouton est relaché, on remet clic_en_cours à False
    
    
def draw() :
    global LARGE,DECAL_X,DECAL_Y,NB_IMAGES
    global FLECHE_JAUNE,FLECHE_ROUGE,JOUEUR
    
    if JOUEUR == 1 :
        fleche = FLECHE_ROUGE
    else :
        fleche = FLECHE_JAUNE
        
    for i in range(NB_IMAGES) :
        # on dessine l'image
        image(fleche, DECAL_X + i*LARGE,DECAL_Y)
        
    affiche_grille(DECAL_X,DECAL_Y+56)
    
    clic()
    
##########################################
####     PROGRAMME PRINCIPAL          ####
##########################################

# initialisation de la grille
GRILLE = creation_grille_vierge()

# largeur/hauteur des images :
LARGE = 40
HAUT = 56
NB_IMAGES = 7

# position des images 
DECAL_X = 10
DECAL_Y = 40

# lancement :
JOUEUR = 1
CLIC_EN_COURS = False

run(preload=preload)
````

````python
# le code complet

##################################
#####   Imports de modules #######
##################################
from module_part4 import *
from p5 import *
from random import randint
import time

def preload():
    """
    Création des images, qui sont toutes variables globales
    """
    global FLECHE_JAUNE, FLECHE_ROUGE, FLECHE_ROUGE_PLEINE, FLECHE_JAUNE_PLEINE, START, VICTOIRE, DEFAITE, NUL
    
    FLECHE_JAUNE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-34-01/flehce_jaune.jpg")
    FLECHE_ROUGE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-19-31/flehce_rouge.jpg")
    FLECHE_ROUGE_PLEINE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-19-35/flehce_rouge_pleine.jpg")
    FLECHE_JAUNE_PLEINE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-18/15-34-05/flehce_jaune_pleine.jpg")
    START = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-19/21-57-45/start.jpg")
    VICTOIRE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-20/1-35-07/victoire.jpg")
    DEFAITE = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-20/1-35-03/defaite.jpg")
    NUL = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/06-20/1-35-17/partie_nulle.jpg")
    
def setup() :
    createCanvas(300, 400) # on cré une fenêtre de 100x100 pixels
    background("#ivory")  # on choisi une couleur pour le fond

def image_cliquee(x,y) :
    """
    IN : x et y : coordonnées du clic
    OUT :
        - si on a cliqué dans la zone des images : la fonction renvoie le numéro de l'image cliquée
        - Sinon : elle renvoie -1
    """
    global LARGE,HAUT,DECAL_X,DECAL_Y,NB_IMAGES
    
    if DECAL_X <= x <= DECAL_X + NB_IMAGES * LARGE and DECAL_Y <= y <= DECAL_Y + HAUT :
        return int(x-DECAL_X)//LARGE
    return -1

def affiche_grille(x,y) :
    """
    Dession une grille constituée de 7 lignes, contenant chacune 7 carrés.
    La couleur des carrés est (0,0,200), leur taille est indiquée par la variable globale LARGE.
    Le premier carré est déssiné aux coordonnées x,y reçues en paramètres.
    """
    global LARGE,HAUT
    
    for ligne in range(6) :
        for colonne in range(7) :
            fill(0,0,200)
            square(x+colonne*LARGE, y+ligne*LARGE, LARGE)
            pion(x+colonne*LARGE , y+ligne*LARGE, ligne, colonne)
    
def pion(x,y,i,j) :
    global GRILLE,LARGE
    centre_x = x + LARGE//2
    centre_y = y + LARGE//2
    if GRILLE[i][j] == 0 : 
        fill("white")
        circle(centre_x,centre_y,LARGE * 3 //4,)
    elif GRILLE[i][j] == 1 : 
        fill("red")
        circle(centre_x,centre_y,LARGE * 3 //4,)
        stroke(200,0,0)
        circle(centre_x,centre_y,LARGE * 3 //5)
        circle(centre_x,centre_y,LARGE * 3//10)
        noStroke()
    else : 
        fill("yellow")
        circle(centre_x,centre_y,LARGE * 3 //4,)
        stroke(200,200,0)
        circle(centre_x,centre_y,LARGE * 3 //5)
        circle(centre_x,centre_y,LARGE * 3//10)
        noStroke()
 

def detect_joueur(x,y) :
    """
    IN : x et y sont les coordonnées du clic
    OUT : Rien mais on utilise les variables golbale ORDI et MODE
    Si un clic sur un jeton est reconnue, la fonction modifie les variables globales ORDI et MODE
    Sinon elle ne fait rien
    """
    
    global ORDI, MODE
    
    # détection d'un clic sur jeton rouge :
    # il faut détecter si l'on a cliqué le jeton 
    # Pour cela on va tester si la distance (au carré) au centre du jeton rouge est inférieure au rayon (au carré) du jeton.
    # le centre du jeton rouge est (272.5 , 170) et son rayon est 20. 
    # La distance (au carré) du clic au centre est donc :
    #                           d_carre = (mouseX - 272.5)**2 + (mouseY - 170)**2
    d_carre = (x-272.5)**2 + (y-170)**2
    if d_carre < 20**2 :
        ORDI = 2
        MODE = "game"
        return
    
    # détection d'un clic sur jeton jaune :
    # faite de même, en adaptant et en sachant que 
    # le centre du jeton jaune est aux coordonnées 272.5,223, avec aussi un rayon de 20.
    d = (x-272.5)**2 + (y-223)**2
    if d < 20**2 :
        ORDI = 1
        MODE = "game"    

def detect_jouer(x,y) :
    """
    IN : x et y sont les coordonnées du clic
    OUT : Rien mais on utilise les variables golbale GRILLE, JOUEUR et MODE
    Si un clic sur un jeton est reconnu, la fonction modifie GRILLLE (via joue_jeton())
    Dans ce cas, elle vérifie ensuite si le joueur a gagné. Si oui, elle modifie MODE à "endgame" sinon elle modifie JOUEUR
    Si on a pas cliqué sur un bouton, la fonction ne fait rien
    """ 
    global GRILLE, JOUEUR, MODE
    num = image_cliquee(x,y)
    if num >= 0 :
        GRILLE = joue_jeton(GRILLE,JOUEUR,num)
        
         # on vérifie si le joueur  à gagné (oui=> MODE endgame, non => changer de joueur)
        if check_victoire(GRILLE,JOUEUR) :
            MODE = "endgame"
        else :
            JOUEUR = 3 - JOUEUR

def boutons() :
    """
    Entrée : RIEN - mais utilise les variables globlales suivantes :
      - JOUEUR
      - 4 images (définies dans preload)
      - paramètres des images : LARGE, DECAL_X, DECAL_Y, NB_IMAGES
    SORTIE : rien : la fonction affiche les images
    """
    global FLECHE_JAUNE, FLECHE_ROUGE, FLECHE_ROUGE_PLEINE, FLECHE_JAUNE_PLEINE
    global JOUEUR
    global LARGE, DECAL_X, DECAL_Y, NB_IMAGES
        
    if JOUEUR == 1 :
        fleche = FLECHE_ROUGE
        fleche_pleine = FLECHE_ROUGE_PLEINE
    else :
        fleche = FLECHE_JAUNE
        fleche_pleine = FLECHE_JAUNE_PLEINE
        
    for i in range(NB_IMAGES) :
        # on dessine l'image
        if colonne_pleine(GRILLE,i) :
            image(fleche_pleine, DECAL_X + i*LARGE,DECAL_Y)
        else :
            image(fleche, DECAL_X + i*LARGE,DECAL_Y)
            


def affiche_tour() :
    """
    IN : rien mais utilise les variables globales JOUEUR et ORDI
    OUT : RIEN - La fonction affiche un message en haut de la fenêtre de jeu
    """
    if JOUEUR != ORDI :
        fill("black")
        textSize(24)
        textAlign(CENTER)
        text("A vous de jouer",150,20)
    else :
        fill("black")
        textSize(24)
        textAlign(CENTER)
        text("L'ordinateur joue...",150,20)

def coup_ordi() :
    """
    IN : rien mais utilise les variables globales GRILLE, JOUEUR et MODE
    OUT : rien, la fonction modifie les variables globales GRILLE, JOUEURS et MODE
    
    La fonction modifie GRILLLE (via joue_jeton()) (avec un timer de 2s)
    Elle vérifie ensuite si le joueur ordinateur a gagné. 
    Si oui, elle modifie MODE à "endgame" sinon elle modifie JOUEUR
    """
    global GRILLE,JOUEUR, FREEZE, MODE
    
    if FREEZE == 0 : # c'est la 1ere exécution de draw() l'ordi doit jouer, mais on ne veux pas qu'il joue de suite...
        FREEZE = time.time() # on retient l'heure de début de son tour de jeu
    elif time.time() > FREEZE + 2 : # 2 secondes sont passées, il joue :
        colonnes = [col for col in range(7) if not colonne_pleine(GRILLE,col)]
        GRILLE = joue_jeton(GRILLE,JOUEUR,colonnes[randint(0,len(colonnes)-1)])
        
        # on vérifie si le joueur (ordinateur)  à gagné (oui=> MODE endgame, non => changer de joueur)
        if check_victoire(GRILLE,JOUEUR) :
            MODE = "endgame"
        else :
            JOUEUR = 3 - JOUEUR
        FREEZE = 0 # et on remet à 0 (il ne rejouera pas car entre temps, on a changé de joueur)

def detect_rejouer(x,y) :
    """
    IN : x et y : coordonnées du clic. De plus, on utilise les variables globales MODE, GRILLE et JOUEUR
    OUT : Rien
    Si on a cliqué le bouton rejoué, la fonction modifie GRILLE MODE et JOUEUR
    Sinon elle ne fait rien
    """
    global MODE,JOUEUR,GRILLE
    if 105 < x < 180 and 56 < y < 75 :
        MODE = "start"
        GRILLE = creation_grille_vierge()
        JOUEUR = 1
        
def clic() :
    global CLIC_EN_COURS
    global JOUEUR,GRILLE,ORDI,MODE
   
    if mouseIsPressed :
        if not CLIC_EN_COURS :          # si un clic n'est pas deja en cours
            CLIC_EN_COURS = True        # alors on met clic_en_cours à True pour bloquer d'autre exécution
            
            if MODE == "start" :
                detect_joueur(mouseX,mouseY)
            elif MODE == "game" :
                detect_jouer(mouseX,mouseY)
            else : # MODE == "engame" :
                detect_rejouer(mouseX,mouseY)
    else :
        CLIC_EN_COURS = False           # le bouton est relaché, on remet clic_en_cours à False
        
def draw() :
    """
    Entrée : RIEN - mais utilise les variables globlales DECAL_X et DECAL_Y
    SORTIE : rien : la fonction affiche appelle les fonctions boutons() et affiche_grille()
    """
    global DECAL_X,DECAL_Y
    global START
    global MODE
    
    clear()
    
    if MODE == "start" :
        image(START,0,120)
    elif MODE == "game" :
        affiche_tour()
        boutons()
        affiche_grille(DECAL_X,DECAL_Y+56)
        if JOUEUR == ORDI : coup_ordi()
        
    else : # MODE = "endgame"
        if JOUEUR == ORDI : 
            image(DEFAITE,0,0)
        else :
            image(VICTOIRE,0,0)
        affiche_grille(DECAL_X,DECAL_Y+56)
    
    clic()
    
####################################################
###                PROGRAMME PRINCIPAL          ####
####################################################

# initialisation de la grille
GRILLE = creation_grille_vierge()

# largeur/hauteur des images :
LARGE = 40
HAUT = 56
NB_IMAGES = 7

# position des images 
DECAL_X = 10
DECAL_Y = 40

# MODE :
MODE = "start"

# lancement :
JOUEUR = 1
CLIC_EN_COURS = False
FREEZE = 0

run(preload=preload)
````
